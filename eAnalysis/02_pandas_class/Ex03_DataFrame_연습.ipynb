#### 행에서 데이타 추출

- 야채, 딸기, 봉투행을 추출하여 매출순으로 정렬하기 (sort_values()이용)
- 바나나행부터 딸기행까지 추출
- 야채부터 딸기행까지 추출

```python
#야채 딸기 봉투 행 출력하는 방법
df1 = df[(df['상품명']=='야채') | (df['상품명']=='딸기')| (df['상품명']=='봉투')]
df1

#그 3가지 행을 매출순으로 정렬하기 #  # 내림차순 : ascending=False 
df2 = df1.sort_values('매출',ascending=False)
df2

#바나나행부터 딸기행까지 추출
df[1:6]

#야채부터 딸기까지
df[2:6]


```
