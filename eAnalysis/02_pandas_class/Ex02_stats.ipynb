```python
# pandas 기술 통계 확인
from pandas import DataFrame

df=DataFrame({'거래처':['A매장','B매장','C매장','A매장','C매장','C매장', 'C매장', 'C매장'],
             '주문량':[100,25, 40, 55, 137, 70, 80, 89],
             '할인율':[0.1, 0.3, 0.03, 0.07, 0.15, 0.2, 0.2, 0.2]})

df
```

### [ 연습]

1. 총합

2. 평균

3. 요약통계량

4. 누적합

5. 주문량 * 할인율

6. 주문량만 누적합 구하기

7. 거래처별 거래 수를 구하고 그래프로 출력 ( 그래프에서 한글이 깨짐 )

    cf) value_counts()

```python
#1번 총합구하기
ds =df.sum()
ds
```

```python
#2번 평균
dm = df.mean()
dm
```

```python
#3요약통계량
dc = df.describe()
dc
```

```python
#4번 누적합
dc = df.cumsum()
dc
```

```python
#5번 주문량 x 할인률
df['주문량*할인율']  = df['주문량'] * df['할인율']
df['주문량*할인율'] 
df
                             
```

```python
#6번 주문량만 누적합 구하기
df['주문누적합'] = df['주문량'].cumsum()
df
```

```python
#7번 거래처별 거래수 그래프까지
dg = df['거래처'].value_counts()
dg.plot()
```
